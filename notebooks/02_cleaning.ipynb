```python
import pandas as pd
import numpy as np
import re

df = pd.read_csv("../data/raw/tweets_raw.csv")
print(f"✅ {len(df)} satır yüklendi")
df.head()
```

```python
# created_at sütununu datetime'a çevir
df["created_at"] = pd.to_datetime(df["created_at"], errors="coerce")

# Sayısal sütunları düzelt
df["likes"]    = pd.to_numeric(df["likes"],    errors="coerce").fillna(0).astype(int)
df["retweets"] = pd.to_numeric(df["retweets"], errors="coerce").fillna(0).astype(int)

print(df.dtypes)
```

```python
def clean_tweet(text):
    text = str(text)
    text = re.sub(r"http\S+|www\S+", "", text)        # URL kaldır
    text = re.sub(r"@\w+", "", text)                   # mention kaldır
    text = re.sub(r"#(\w+)", r"\1", text)              # # işaretini kaldır
    text = re.sub(r"[^\w\s]", " ", text, flags=re.UNICODE)  # noktalama
    text = re.sub(r"\d+", "", text)                    # rakamlar
    text = text.lower().strip()
    text = re.sub(r"\s+", " ", text)                   # fazla boşluk
    return text

df["clean_text"] = df["text"].apply(clean_tweet)

# Duplicate kontrol
before = len(df)
df = df.drop_duplicates(subset=["text"]).reset_index(drop=True)
print(f"Duplicate kaldırıldı: {before - len(df)} satır")
print(f"Kalan: {len(df)} satır")
```

```python
import os
os.makedirs("../data/processed", exist_ok=True)

df.to_csv("../data/processed/tweets_cleaned.csv", index=False, encoding="utf-8-sig")
print(f"✅ Kaydedildi: data/processed/tweets_cleaned.csv ({len(df)} satır)")
```
