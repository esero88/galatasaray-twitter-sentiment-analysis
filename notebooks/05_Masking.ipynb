```python
# CSV'yi oku
df = pd.read_csv("C:/Users/Eser/galatasaray-twitter-sentiment-analysis/data/processed/tweets_sentiment.csv")
```

```python
# Her benzersiz username'e sıralı, tutarlı bir takma ad ata
unique_users = df['username'].unique()
mapping = {user: f"user_{i+1:04d}" for i, user in enumerate(unique_users)}
```

```python
# Orijinal username sütununu maskeli hale ile değiştir
df['username'] = df['username'].map(mapping)
```

```python
# Maskelenmiş CSV'yi doğru klasöre kaydet
df.to_csv("C:/Users/Eser/galatasaray-twitter-sentiment-analysis/data/processed/tweets_sentiment_masked.csv", index=False)
```
