```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
```

```python
# Google Sheets'ten veri yükle
SHEET_ID = "114yXOgNpEeLxmVSNSk4wEviOtewZ0IEUz8RYfp_CiPg"

url = f"https://docs.google.com/spreadsheets/d/{SHEET_ID}/export?format=csv&gid=0"
df = pd.read_csv(url)

print(f"✅ {len(df)} satır, {len(df.columns)} sütun yüklendi")
```

```python
print(f"Öncesi: {len(df)} satır")

df = df.dropna(how="all")           # tamamen boş satırları at
df = df.dropna(subset=["text"])     # text'i boş olanları at
df = df.reset_index(drop=True)

print(f"Sonrası: {len(df)} satır")
```

```python
import os
os.makedirs("../data/raw", exist_ok=True)

df.to_csv("../data/raw/tweets_raw.csv", index=False, encoding="utf-8-sig")
print(f"✅ Kaydedildi: ../data/raw/tweets_raw.csv ({len(df)} satır)")
```
